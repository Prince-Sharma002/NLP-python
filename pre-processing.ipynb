```python
import spacy

nlp = spacy.load("en_core_web_sm")


doc = nlp("tata and Reliance is a known company in india which worth ₹20000 net worth")

for token in doc:
    print(token.text, "|" , token.pos_, "|" ,  token.dep_  , "|" , token.lemma_,  "|" , token.tag    )


from spacy import displacy

for ent in doc.ents:
    print(ent.text, "|" , ent.label_ )

displacy.render(doc , style="ent")


```

```python
# bow


import pandas as pd
import numpy as np

df = pd.read_csv("spam.csv")
df.head()

```

```python
df.Category.value_counts()
```

```python
def func(x):
    if(x == "spam"):
        return 1
    return 0

df['spam'] = df["Category"].apply(func)
df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.Message, df.spam , test_size=0.2 )

```

```python
from sklearn.feature_extraction.text import CountVectorizer

v = CountVectorizer()

X_train_cv =  v.fit_transform( X_train.values )

X_train_np = X_train_cv.toarray()

np.where( X_train_np[0] != 0)

X_train_np[0][951]
```

```python
# creating model

from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train_cv , y_train)

X_test_cv = v.fit_transform(X_test.values)

from sklearn.metrics import classification_report

y_pred = model.predict(X_test_cv)

# print(classification_report(y_test, y_pred))
```
